```python
from openpyxl import Workbook, load_workbook
from openpyxl.utils import get_column_letter
import pandas as pd
import numpy as np
#By Steven Hua
#For the Getting to Know U VIP
#Research Space Utilization
#V 1.0


df = load_workbook("test worksheet.xlsx", data_only=True)
ws = df.active

#Creates a empty dataframe with Column EMPLID which is the data we want
finalDF = pd.DataFrame(columns=["Emplid"])

#This loop looks at if the VLOOKUP sheet is true, if true we append it to a separate spreadsheet
for row in range(3, 953):
    booleanVal = get_column_letter(7)
    empID = get_column_letter(1)
    if(ws[booleanVal + str(row)].value == True):
        
        finalDF = finalDF.append({"Emplid":ws[empID + str(row)].value}, ignore_index = True)

print(finalDF)

#Exports our combined worksheet to an excel file
finalDF.to_excel("combined.xlsx")
```
